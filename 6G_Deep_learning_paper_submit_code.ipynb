```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import class_weight

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv1D, LSTM, Flatten, Concatenate, Dropout, MultiHeadAttention, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, Callback
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (confusion_matrix, classification_report, roc_curve, auc,
                             accuracy_score, precision_score, recall_score, f1_score, roc_auc_score)

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

# --- Set Random Seed for Reproducibility ---
tf.random.set_seed(42)
np.random.seed(42)

# --- Load Dataset ---
df = pd.read_csv("medical_resource_allocation_binary_dataset.csv")

# --- Dataset Diagnostics ---
print("Dataset Shape:", df.shape)
print("Class Distribution:\n", df['Resource_Utilization_Efficiency'].value_counts(normalize=True))
print("Missing Values:\n", df.isnull().sum())

# --- Add Hypothetical 6G Features ---
df['6G_Signal_Strength'] = np.random.uniform(0, 1, df.shape[0])
df['6G_Impact_Factor'] = np.random.uniform(0, 1, df.shape[0])

# --- Encode Categorical Columns ---
label_encoders = {}
for col in ['Gender', 'Health_Status', 'Disease_Type', 'Resource_Needed', 'Sensor_ID', 'Resource_ID']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# --- Define Features and Target ---
target_column = 'Resource_Utilization_Efficiency'
X = df.drop(columns=[target_column, 'Patient_ID'])
y = df[target_column]

# --- Normalize Features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Split 6G and Main Features ---
X_main = X_scaled[:, :-2].reshape((X_scaled.shape[0], X_scaled.shape[1] - 2, 1))  # Reshape for Conv1D
X_6G = X_scaled[:, -2:]

# --- Train-Test Split (Stratified) ---
X_train_main, X_test_main, X_train_6G, X_test_6G, y_train, y_test = train_test_split(
    X_main, X_6G, y, test_size=0.2, stratify=y, random_state=42)

# --- Compute Class Weights ---
class_weights = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = {i: class_weights[i] for i in range(len(class_weights))}
print("Class Weights:", class_weights_dict)

# ===============================
# === Fixed Hybrid Deep Learning Model
# ===============================
class LearningRateWarmup(Callback):
    def __init__(self, warmup_epochs=5, start_lr=1e-5, target_lr=5e-4):
        super(LearningRateWarmup, self).__init__()
        self.warmup_epochs = warmup_epochs
        self.start_lr = start_lr
        self.target_lr = target_lr

    def on_epoch_begin(self, epoch, logs=None):
        if epoch < self.warmup_epochs:
            lr = self.start_lr + (self.target_lr - self.start_lr) * (epoch / self.warmup_epochs)
            self.model.optimizer.learning_rate.assign(lr)
            print(f"Epoch {epoch+1}: Setting learning rate to {lr:.6f}")

def create_hybrid_model():
    input_main = Input(shape=(X_train_main.shape[1], 1))
    input_6G = Input(shape=(2,))

    x = Conv1D(filters=32, kernel_size=3, activation='relu', kernel_regularizer=l2(0.0005))(input_main)
    x = BatchNormalization()(x)
    x = LSTM(32, return_sequences=True, kernel_regularizer=l2(0.0005))(x)
    x = Dropout(0.25)(x)
    x = MultiHeadAttention(num_heads=4, key_dim=32)(x, x)
    x = Flatten()(x)

    merged = Concatenate()([x, input_6G])
    merged = Dense(128, activation='relu', kernel_regularizer=l2(0.0005))(merged)
    merged = Dropout(0.25)(merged)
    output = Dense(1, activation='sigmoid')(merged)

    model = Model(inputs=[input_main, input_6G], outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# --- Create and Summarize Model ---
hybrid_model = create_hybrid_model()
hybrid_model.summary()

# --- Callbacks ---
early_stop = EarlyStopping(monitor='val_loss', patience=7, min_delta=0.0005, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)
lr_warmup = LearningRateWarmup(warmup_epochs=5, start_lr=1e-5, target_lr=5e-4)

# --- Train Hybrid Model ---
history = hybrid_model.fit(
    [X_train_main, X_train_6G], y_train,
    validation_data=([X_test_main, X_test_6G], y_test),
    epochs=30,
    batch_size=32,
    callbacks=[lr_warmup, early_stop, reduce_lr],
    class_weight=class_weights_dict,
    verbose=1
)

# --- Evaluate Hybrid Model ---
y_pred_dl = (hybrid_model.predict([X_test_main, X_test_6G]) > 0.5).astype("int32")
y_prob_dl = hybrid_model.predict([X_test_main, X_test_6G])

acc_train_hybrid = history.history['accuracy'][-1]
acc_test_hybrid = accuracy_score(y_test, y_pred_dl)
prec_hybrid = precision_score(y_test, y_pred_dl)
rec_hybrid = recall_score(y_test, y_pred_dl)
f1_hybrid = f1_score(y_test, y_pred_dl)
roc_hybrid = roc_auc_score(y_test, y_prob_dl)

# --- Confusion Matrix - Hybrid ---
print("\n--- Hybrid Deep Learning Model ---")
print("Train Accuracy:", f"{acc_train_hybrid:.4f}")
print("Test Accuracy:", f"{acc_test_hybrid:.4f}")
print(classification_report(y_test, y_pred_dl))
cm_dl = confusion_matrix(y_test, y_pred_dl)
sns.heatmap(cm_dl, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Hybrid DL Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# --- Graphs for Accuracy and Loss ---
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ================================
# === Traditional ML Models
# ================================
X_train_ml, X_test_ml, y_train_ml, y_test_ml = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42)

models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(kernel='rbf', probability=True, random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

results = []

def train_ml_model(name, model, X_train, y_train, X_test, y_test, results):
    model.fit(X_train, y_train)

    y_pred_test = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    y_prob = model.predict_proba(X_test)[:, 1]

    acc_train = accuracy_score(y_train, y_pred_train)
    acc_test = accuracy_score(y_test, y_pred_test)
    prec = precision_score(y_test, y_pred_test)
    rec = recall_score(y_test, y_pred_test)
    f1 = f1_score(y_test, y_pred_test)
    roc = roc_auc_score(y_test, y_prob)

    results.append([name, acc_train, acc_test, prec, rec, f1, roc])

    print(f"\n--- {name} ---")
    print("Train Accuracy:", f"{acc_train:.4f}")
    print("Test Accuracy :", f"{acc_test:.4f}")
    print(classification_report(y_test, y_pred_test))

    cm = confusion_matrix(y_test, y_pred_test)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

for name, model in models.items():
    train_ml_model(name, model, X_train_ml, y_train_ml, X_test_ml, y_test_ml, results)

# Add Hybrid Model Result
results.append(["Hybrid Deep Learning", acc_train_hybrid, acc_test_hybrid, prec_hybrid, rec_hybrid, f1_hybrid, roc_hybrid])

# === Results Summary Table ===
df_results = pd.DataFrame(results, columns=["Model", "Train Accuracy", "Test Accuracy", "Precision", "Recall", "F1 Score", "AUC"])
print("\n=== Model Performance Summary ===")
print(df_results.to_string(index=False))

# === ROC Curve Comparison ===
def plot_roc(models, X_test, y_test, hybrid_model=None, X_test_main=None, X_test_6G=None):
    plt.figure(figsize=(8, 6))
    for name, model in models.items():
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        plt.plot(fpr, tpr, label=f"{name} (AUC = {auc(fpr, tpr):.2f})")

    if hybrid_model:
        y_prob_hybrid = hybrid_model.predict([X_test_main, X_test_6G])
        fpr, tpr, _ = roc_curve(y_test, y_prob_hybrid)
        plt.plot(fpr, tpr, label=f"Hybrid DL (AUC = {auc(fpr, tpr):.2f})", linestyle='--')

    plt.plot([0, 1], [0, 1], 'k--', label="Random")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve Comparison")
    plt.legend()
    plt.show()

plot_roc(models, X_test_ml, y_test_ml, hybrid_model, X_test_main, X_test_6G)

# === Bar Charts for Metrics ===
metrics = ['Train Accuracy', 'Test Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC']
for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.barplot(x="Model", y=metric, data=df_results, palette="magma")
    plt.ylim(0, 1)
    plt.title(f"Model {metric} Comparison")
    plt.ylabel(metric)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# === Feature Importance (RF) ===
importances = models["Random Forest"].feature_importances_
feature_names = df.drop(columns=[target_column, 'Patient_ID']).columns

plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=feature_names, palette="viridis")
plt.title("Random Forest Feature Importances")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, Concatenate, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (confusion_matrix, classification_report, roc_curve, auc,
                             accuracy_score, precision_score, recall_score, f1_score, roc_auc_score)
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

# --- Set Random Seed for Reproducibility ---
tf.random.set_seed(42)
np.random.seed(42)

# --- Load Dataset ---
df = pd.read_csv("medical_resource_allocation_binary_dataset.csv")

# --- Dataset Diagnostics ---
print("Dataset Shape:", df.shape)
print("Class Distribution:\n", df['Resource_Utilization_Efficiency'].value_counts(normalize=True))
print("Missing Values:\n", df.isnull().sum())

# --- Add Hypothetical 6G Features ---
df['6G_Signal_Strength'] = np.random.uniform(0, 1, df.shape[0])
df['6G_Impact_Factor'] = np.random.uniform(0, 1, df.shape[0])

# --- Encode Categorical Columns ---
label_encoders = {}
for col in ['Gender', 'Health_Status', 'Disease_Type', 'Resource_Needed', 'Sensor_ID', 'Resource_ID']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# --- Define Features and Target ---
target_column = 'Resource_Utilization_Efficiency'
X = df.drop(columns=[target_column, 'Patient_ID'])
y = df[target_column]

# --- Normalize Features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Split 6G and Main Features ---
X_main = X_scaled[:, :-2]  # Main features (tabular, no reshaping)
X_6G = X_scaled[:, -2:]    # 6G features

# --- Train-Test Split (Stratified) ---
X_train_main, X_test_main, X_train_6G, X_test_6G, y_train, y_test = train_test_split(
    X_main, X_6G, y, test_size=0.2, stratify=y, random_state=42)

# ===============================
# === Simplified Hybrid Deep Learning Model
# ===============================
def create_hybrid_model():
    input_main = Input(shape=(X_train_main.shape[1],))
    input_6G = Input(shape=(2,))

    # Feedforward layers for main features
    x = Dense(64, activation='relu', kernel_regularizer=l2(0.0001))(input_main)
    x = BatchNormalization()(x)
    x = Dropout(0.2)(x)
    x = Dense(32, activation='relu', kernel_regularizer=l2(0.0001))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.2)(x)

    # Concatenate with 6G features
    merged = Concatenate()([x, input_6G])
    merged = Dense(16, activation='relu', kernel_regularizer=l2(0.0001))(merged)
    merged = Dropout(0.2)(merged)
    output = Dense(1, activation='sigmoid')(merged)

    model = Model(inputs=[input_main, input_6G], outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# --- Callbacks ---
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)

# --- Train Hybrid Model ---
hybrid_model = create_hybrid_model()
history = hybrid_model.fit(
    [X_train_main, X_train_6G], y_train,
    validation_data=([X_test_main, X_test_6G], y_test),
    epochs=20,  # Increased epochs
    batch_size=32,  # Larger batch size
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

# --- Evaluate Hybrid Model ---
y_pred_dl = (hybrid_model.predict([X_test_main, X_test_6G]) > 0.5).astype("int32")
y_prob_dl = hybrid_model.predict([X_test_main, X_test_6G])

acc_train_hybrid = history.history['accuracy'][-1]
acc_test_hybrid = accuracy_score(y_test, y_pred_dl)
prec_hybrid = precision_score(y_test, y_pred_dl)
rec_hybrid = recall_score(y_test, y_pred_dl)
f1_hybrid = f1_score(y_test, y_pred_dl)
roc_hybrid = roc_auc_score(y_test, y_prob_dl)

# --- Confusion Matrix - Hybrid ---
print("\n--- Hybrid Deep Learning Model ---")
print("Train Accuracy:", f"{acc_train_hybrid:.4f}")
print("Test Accuracy:", f"{acc_test_hybrid:.4f}")
print(classification_report(y_test, y_pred_dl))
cm_dl = confusion_matrix(y_test, y_pred_dl)
sns.heatmap(cm_dl, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Hybrid DL Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# --- Graphs for Accuracy and Loss ---
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ================================
# === Traditional ML Models
# ================================
X_train_ml, X_test_ml, y_train_ml, y_test_ml = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42)

models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(kernel='rbf', probability=True, random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

results = []

def train_ml_model(name, model, X_train, y_train, X_test, y_test, results):
    model.fit(X_train, y_train)
    y_pred_test = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    y_prob = model.predict_proba(X_test)[:, 1]

    acc_train = accuracy_score(y_train, y_pred_train)
    acc_test = accuracy_score(y_test, y_pred_test)
    prec = precision_score(y_test, y_pred_test)
    rec = recall_score(y_test, y_pred_test)
    f1 = f1_score(y_test, y_pred_test)
    roc = roc_auc_score(y_test, y_prob)

    results.append([name, acc_train, acc_test, prec, rec, f1, roc])

    print(f"\n--- {name} ---")
    print("Train Accuracy:", f"{acc_train:.4f}")
    print("Test Accuracy :", f"{acc_test:.4f}")
    print(classification_report(y_test, y_pred_test))

    cm = confusion_matrix(y_test, y_pred_test)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

for name, model in models.items():
    train_ml_model(name, model, X_train_ml, y_train_ml, X_test_ml, y_test_ml, results)

# Add Hybrid Model Result
results.append(["Hybrid Deep Learning", acc_train_hybrid, acc_test_hybrid, prec_hybrid, rec_hybrid, f1_hybrid, roc_hybrid])

# === Results Summary Table ===
df_results = pd.DataFrame(results, columns=["Model", "Train Accuracy", "Test Accuracy", "Precision", "Recall", "F1 Score", "AUC"])
print("\n=== Model Performance Summary ===")
print(df_results.to_string(index=False))

# === ROC Curve Comparison ===
def plot_roc(models, X_test, y_test, hybrid_model=None, X_test_main=None, X_test_6G=None):
    plt.figure(figsize=(8, 6))
    for name, model in models.items():
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        plt.plot(fpr, tpr, label=f"{name} (AUC = {auc(fpr, tpr):.2f})")

    if hybrid_model:
        y_prob_hybrid = hybrid_model.predict([X_test_main, X_test_6G])
        fpr, tpr, _ = roc_curve(y_test, y_prob_hybrid)
        plt.plot(fpr, tpr, label=f"Hybrid DL (AUC = {auc(fpr, tpr):.2f})", linestyle='--')

    plt.plot([0, 1], [0, 1], 'k--', label="Random")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve Comparison")
    plt.legend()
    plt.show()

plot_roc(models, X_test_ml, y_test_ml, hybrid_model, X_test_main, X_test_6G)

# === Bar Charts for Metrics ===
metrics = ['Train Accuracy', 'Test Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC']
for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.barplot(x="Model", y=metric, data=df_results, palette="magma")
    plt.ylim(0, 1)
    plt.title(f"Model {metric} Comparison")
    plt.ylabel(metric)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# === Feature Importance (RF) ===
importances = models["Random Forest"].feature_importances_
feature_names = df.drop(columns=[target_column, 'Patient_ID']).columns

plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=feature_names, palette="viridis")
plt.title("Random Forest Feature Importances")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()
```
